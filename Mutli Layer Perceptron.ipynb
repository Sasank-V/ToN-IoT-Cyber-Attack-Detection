### Binary Classification
- Attack(1) or No Attack(0)

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
```

```python
X = df.drop(columns=["Label","Attack"])
y = df["Label"]
```

```python
np.isinf(X).sum()
X.replace([np.inf, -np.inf], np.nan, inplace=True)
X.fillna(X.median(), inplace=True)
```

```python
X_sample = X.sample(10000, random_state=1010)  # 10k rows
y_sample = y.loc[X_sample.index]
scaler = StandardScaler()
X_train, X_test, y_train, y_test = train_test_split(
    X_sample, y_sample, test_size=0.2, random_state=1010, stratify=y_sample
)
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

- Activation Function: Relu

```python
mlp = MLPClassifier(hidden_layer_sizes=(64,32),max_iter=200,random_state=1010,activation='relu',solver='adam')
mlp.fit(X_train,y_train)

y_pred = mlp.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)

print(f"Accuracy: {accuracy}")
```

- Activation Function: tanh

```python
mlp = MLPClassifier(hidden_layer_sizes=(64,32),max_iter=200,random_state=1010,activation='tanh',solver='adam')
mlp.fit(X_train,y_train)

y_pred = mlp.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)

print(f"Accuracy: {accuracy}")
```

- Activation Function: logistic (sigmoid)

```python
mlp = MLPClassifier(hidden_layer_sizes=(64,32),max_iter=500,random_state=1010,activation='logistic',solver='adam')
mlp.fit(X_train,y_train)

y_pred = mlp.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)

print(f"Accuracy: {accuracy}")
```

### Multiclass Classification ( 0 - 8 )

```python
from sklearn.preprocessing import LabelEncoder

# X = features
X = df.drop(columns=['Label', 'Attack'])
# y = Attack column
y = df['Attack']

# Encode categorical labels to numbers
le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
import numpy as np

X.replace([np.inf, -np.inf], np.nan, inplace=True)
X.fillna(X.median(), inplace=True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report

mlp = MLPClassifier(
    hidden_layer_sizes=(128, 64),
    activation='relu',
    solver='adam',
    max_iter=300,
    random_state=42
)

# Train
mlp.fit(X_train, y_train)

# Predict
y_pred = mlp.predict(X_test)

# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=le.classes_))
```

# Conclusion

1. **Binary Classification (Label: 0 or 1)**

   * MLP with two hidden layers `(64,32)` performs **very well**:

     * ReLU: 97.45%
     * Tanh: 97.25%
     * Sigmoid: 96.8%
   * All three activations give high accuracy, showing that the model can reliably distinguish **normal vs attack traffic**.
   * ReLU slightly outperforms the others, likely due to better gradient flow and faster convergence.

2. **Multi-class Classification (Attack types)**

   * MLP struggles with **specific attack types**, achieving only \~46% accuracy.
   * Metrics reveal:

     * Majority classes (e.g., `Benign`, `xss`) dominate predictions.
     * Minority classes (e.g., `ddos`, `dos`, `mitm`) are almost never predicted → **very low recall and F1-score**.
   * Challenges arise from **class imbalance**, very few samples for some attacks, and limited feature separation for multiple attack types.

## Justification

* MLP works well for **binary detection** because the classes are clearly separable.
* For **multi-class detection**, the model struggles without additional strategies like:

  * Class balancing (oversampling minority classes)
  * Feature engineering to better separate attack types
  * Larger/deeper networks or alternative models for complex patterns.

Overall, the pre-defined MLP demonstrates **strong capability for binary attack detection** but highlights **limitations for multi-class classification** on imbalanced IoT/CICIDS datasets.
